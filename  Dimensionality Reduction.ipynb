```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data.shape
```

```python
X = data.drop('Species', axis = 1)
X.head()
```

```python
X.shape
```

```python
# Step 1: Apply PCA on the indepedent features (varaiables) X
```

```python
pca = PCA()
principalComponents = pca.fit_transform(X)
print(principalComponents)
```

```python
principalComponents.shape
```

```python
np.round(principalComponents,2)
```

```python
# Step 2 : Check for the variance ratio of the individual principal components
```

```python
np.round(pca.explained_variance_ratio_,2)
```

```python
PC1_vr = 0.92461621
PC2_vr = 0.05301557
np.round(PC1_vr + PC2_vr,2)
```

```python
PC3_vr = 0.01718514 
PC4_vr = 0.00518309
np.round(PC3_vr + PC4_vr,2)
```

```python
# Step 3 : Decide the number of components that the project needs and recreate the PCA model
# and the principal components
```

```python
# Create a PCA model for 2 components
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(X)
print(principalComponents)
```

```python
# By losing 2 coulmns we are still able to retain 98% information of the dataset
```

```python
# Step 4 : Create a new dataframe with these Principal components and the labels (y)
```

```python
df_iris_transformed = pd.DataFrame(principalComponents, columns=['PC1', 'PC2'])
df_iris_transformed
```

```python
df_iris_transformed['Species'] = data['Species']
df_iris_transformed.head()
```

```python
df_iris_transformed.shape
```

```python
df_iris_transformed['Species'].value_counts()
```

```python
# Step 5: Create a Logistic Regression model on this dataset
```

```python
X = df_iris_transformed.drop('Species', axis = 1)
y = df_iris_transformed.Species
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
```

```python
# A user has collected the following values of the Iris flower and wants the model to predict its class
# SL = 5.1, SW = 3.5, PL = 1.4, PW = 0.2
```
